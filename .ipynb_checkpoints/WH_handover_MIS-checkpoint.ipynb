```python
#by SQL extracting base month data
bqclient = bigquery.Client()

query_string = f"""SELECT
  a.invoiceno as handover_Invoice,
  EXTRACT(hour FROM a.handoverdate_time) AS handover_hour,
  CAST(a.handoverdate_time AS date) handover_date,
  CAST(a.boxcount AS int) AS Handover_Boxcount,
  a.volumetricweight as handover_VolWT,
  b.hub_city AS hubname
FROM (
  SELECT
    invoiceno,
    currenthubid,
    COALESCE(zoneinscandate,hubinscandate) AS handoverdate_time,
    mpscount AS boxcount,
    valweight AS volumetricweight
  FROM
    fcdatalake.ftldb.tbl_consignmentmaster
  WHERE
    current_vehicle_id IS NULL
    AND currenthubid IS NOT NULL) AS a
INNER JOIN (
  SELECT
    hub_id,
    CASE
    WHEN hub_city IN ('BANGALORE') THEN 'Bangalore'
    ELSE hub_city
    END AS 
    hub_city,
    
  FROM
    fcdatalake.ftldb.ftl_tbl_hubmaster #where LOWER(hub_city) in ('pune','bangalore','luhari','rajpura','Kolkata')
    ) AS b
ON
  a.currenthubid=b.hub_id
WHERE
  CAST(a.handoverdate_time as date)=date_sub(current_date(), INTERVAL 1 day)"""

df1 = (bqclient.query(query_string).result().to_dataframe(create_bqstorage_client=True))
print(df1.shape)
df1 = df1.sort_values(by='handover_hour')
```

```python
pivot_table = pd.pivot_table(df1, values=['Handover_Boxcount'], index=['handover_hour'], columns=['hubname'], aggfunc='sum', margins=True, margins_name='Total')
pivot_table = pivot_table.fillna(0)
pivot_table = pivot_table.reset_index()
pivot_table = pivot_table[pivot_table['handover_hour'] != 'Total']
pivot_table
```

```python
#by SQL extracting base month data
bqclient = bigquery.Client()

query_string = f"""SELECT
  a.invoiceno AS Total_Invoice,
  EXTRACT(hour FROM a.shipout_timestamp) AS Load_Generated_Hour,
  CAST(a.shipout_timestamp AS date) AS shipout_date,
  CAST(a.boxcount AS int) AS Total_Boxcount,
  a.volumetricweight AS Total_VolWT,
  b.warehousename AS warehousename
FROM (
  SELECT
    invoiceno,
    warehouseid,
    TIMESTAMP(shipoutdate_time) AS shipout_timestamp,
    mpscount AS boxcount,
    valweight AS volumetricweight
  FROM
    fcdatalake.ftldb.tbl_consignmentmaster) AS a
INNER JOIN (
SELECT
    whid,
    CASE  
    WHEN warehousename IN ('Bangalore3_Ext','BLR') THEN 'Bangalore'
    WHEN warehousename IN ('LUHARI_Ext', 'GGN') THEN 'Luhari'
    WHEN warehousename IN ('PNQ','PNQ_Ext') THEN 'Pune'
    WHEN warehousename IN ('Chandigarh_Ext', 'Chandigarh') THEN 'Rajpura'
    WHEN warehousename IN ('KOLKATA_Ext', 'KOL') THEN 'Kolkata'
    WHEN warehousename IN ('AMD','AMD_Ext') THEN 'Ahmedabad'
    ELSE warehousename
  END AS warehousename,
  FROM `fcdatalake.ftldb.ftl_tbl_warehousemaster` WHERE warehousename in ('KOL','AMD','AMD_Ext',
'KOLKATA_Ext','Bangalore3_Ext', 'BLR','LUHARI_Ext', 'GGN','PNQ', 'PNQ_Ext','Chandigarh_Ext', 'Chandigarh')) AS b
ON
  a.warehouseid=b.whid
WHERE
  CAST(a.shipout_timestamp as date)=date_sub(current_date(), INTERVAL 1 day)"""

df2 = (bqclient.query(query_string).result().to_dataframe(create_bqstorage_client=True))
print(df2.shape)
df2 = df2.sort_values(by='Load_Generated_Hour')
```

```python
pivot_table_overallload = pd.pivot_table(df2, values=['Total_Boxcount'], index=['Load_Generated_Hour'], columns=['warehousename'], aggfunc='sum', margins=True, margins_name='Total')
pivot_table_overallload = pivot_table_overallload.fillna(0)
pivot_table_overallload = pivot_table_overallload.reset_index()
pivot_table_overallload = pivot_table_overallload[pivot_table_overallload['Load_Generated_Hour'] != 'Total']
pivot_table_overallload
```

```python
merged_pivot_table = pd.merge(pivot_table_overallload, pivot_table, how='outer', left_on='Load_Generated_Hour', right_on='handover_hour')
merged_pivot_table['Hours'] = merged_pivot_table['Load_Generated_Hour'].combine_first(merged_pivot_table['handover_hour'])
merged_pivot_table = merged_pivot_table.sort_values(by='Hours')
merged_pivot_table = merged_pivot_table.drop((['handover_hour','Load_Generated_Hour']), axis=1, level=0)
merged_pivot_table = merged_pivot_table.set_index('Hours')
merged_pivot_table = merged_pivot_table.fillna(0)
merged_pivot_table.loc['Total'] = merged_pivot_table.sum()
merged_pivot_table
```
